### Data Pre-processing

```python
# Categorical  : One hot encoding
# Numerical : Scaling
```

```python
df
```

```python

df_cat = df [['Item_Fat_Content', 'Outlet_Size', 'Outlet_Location_Type','Outlet_Type' ]]
df_num = df[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Age']]
```

```python
df_cat
```

```python
df['Item_Type'].unique()
```

```python
# Apply One hot encoding to categorical features.

df_cat_dummies = pd.get_dummies(df_cat)

df_cat_dummies
```

```python
# Apply scaling for Numerical features.
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MM = MinMaxScaler()
```

```python
df_num_mm = MM.fit_transform(df_num)
```

```python
df_num_mm = pd.DataFrame(df_num_mm, columns =df_num.columns , index = df_num.index)
```

```python
df_num_mm
```

```python
# Concatinate df_cat_dummies and df_num_mm
```

```python
final = pd.concat([df_num_mm,df_cat_dummies], axis = 1)
```

```python
final
```

```python
# 1. Dependant and Indepedant features.
# 2. Split Data
# 3. Train Data
# 4. Predict Outcomes
# 5. Evaluate ML Model
```

```python
df.columns
```

```python
# 1. Dependent and Independent features
```

```python
X = final                       # Input
Y = df['Item_Outlet_Sales']     # Output
```

```python
# 2. Split Data  80:20
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.20, random_state= 123)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_train, Y_train)
```

```python
# Predict the Outcomes :
```

```python
LR_pred = LR.predict(X_test)
```

```python
LR_pred [:5].round(2)
```

```python
Y_test [:5].round(1)
```

```python
#  Evaluate ML Model
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("Mean Absolute Error : ", mean_absolute_error(Y_test, LR_pred))
```

```python
print(" MSE : ", mean_squared_error(Y_test, LR_pred))
```

```python
r2_score (Y_test, LR_pred)
```
